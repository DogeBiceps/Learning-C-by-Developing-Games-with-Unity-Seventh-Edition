```python
%matplotlib inline
import numpy as np
import json
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
with open("accel_data.json") as f:
    accel_data = json.load(f)

    df = pd.DataFrame(accel_data['accelRecords'])
```

```python
acc_df = pd.json_normalize(df['accel'])
acc_df['time'] = df['time']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(acc_df['time'], acc_df['x'], label='X')
plt.plot(acc_df['time'], acc_df['y'], label='Y')
plt.plot(acc_df['time'], acc_df['z'], label='Z')
plt.title("Accelerometer Data Over Time")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration")
plt.legend()
plt.grid(True)
plt.show()
```

```python
with open("touch_data.json") as f:
    touch_data = json.load(f)
```

```python
touch_df = pd.DataFrame(touch_data['touchRecords'])
```

```python
plt.figure(figsize=(6, 8))
plt.scatter(touch_df['x'], touch_df['y'], c=touch_df['time'], cmap='viridis')
plt.title("Touch Points (Screen Coordinates)")
plt.xlabel("X")
plt.ylabel("Y")
plt.gca().invert_yaxis()
plt.colorbar(label='Time (s)')
plt.grid(True)
plt.show()
```
